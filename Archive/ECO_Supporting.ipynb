```python
def txt_list(parent,children,out_txt, num, letter,n):
    if len(children) == 0:
#         print(parent, 'part without children')
        if num < n + 1:
            num += 1

        for par in parent:
            out_txt.append(f'{num}{letter}. {par[2]} ({par[4]}) is {par[5]} to -{par[6]}')
        out_txt.append('')
        if letter == 'z':
            letter = 'a'
        else:
            letters = string.ascii_letters
            letter = letters[letters.index(letter) +1] 
    elif len(parent) == 1:
        if num < n + 1:
            num += 1
        out_txt.append(f'{num}{letter}. {parent[0][2]} ({parent[0][4]}) is {parent[0][5]} to -{parent[0][6]} for below BOM change:') 
        for child in children:
            out_txt.append(f'--> {child[2]} ({child[4].strip()}) is {child[5]} to -{child[6]}')
        out_txt.append('')
        if letter == 'z':
            letter = 'a'
        else:
            letters = string.ascii_letters
            letter = letters[letters.index(letter) +1] 
    else: #if len(children) == 1: 
        if num < n + 1:
            num += 1
        out_txt.append(f'{num}{letter}. Below parent changed to reflect {children[0][2]} ({children[0][4].strip()}) is {children[0][5]} to -{children[0][6]}:') 
        for par in parent:
            out_txt.append(f'--> {par[2]} ({par[4].strip()}) is {par[5]} to -{par[6]}') 
        out_txt.append('')
        if letter == 'z':
            letter = 'a'
        else:
            letters = string.ascii_letters
            letter = letters[letters.index(letter) +1] 
    return out_txt, num, letter
```

```python
def find_par(child_lst, par_lst, lst_items):
    for child in child_lst:
        loop_index = child[8]
        child_lev = int(child[1])
        child_req = child[0]
        while loop_index + 1 < len(lst_items):

            loop_index = loop_index + 1
            if int(lst_items[loop_index][1]) >= child_lev:
                break

            if int(lst_items[loop_index][1]) == child_lev - 1\
            and lst_items[loop_index][0] == child_req and lst_items[loop_index][7] != 'done':

                lst_items[loop_index][7] = 'done'
                if not any(str(lst_items[loop_index][2]) in par for par in par_lst): #lst_items[m] not in child_lst:
                    par_lst.append(lst_items[loop_index])
    return par_lst
```

```python
# extract and collect from pdf file
def extract_pdf(file_path, file, pn_input,des_input,key_check):
#     print('file_path:',file_path, file, pn_input)
#     print('des_input',des_input)
    red_line = False
    pn_drw = file_path + '/' + file
    doc: Document = fitz.open(pn_drw)
    file_name_result = ''
    pn_result = ''
    rev_result = ''
    des_result = ''
    key_check_result = ''
    pn = pn_input[:11]
    pn_key = pn_input[:7]
    pn_dash = pn_input[8:11]
    pn_rev = pn_input[12:14] #ws.cell(row=i, column=2).value.strip()[12:14]
    if (pn_dash not in file[7:12])\
    or (pn_rev not in file[9:15]):
        file_name_result = 'File name not match ' + '(' + file + ')'
    page: Page = doc[0]
    rect_pn = Rect(page.rect.width-260, page.rect.height-64,\
        page.rect.width-63, page.rect.height-48)
    pn_text = page.get_textbox(rect_pn).rstrip().replace('\n','')
    rect_pn_oms = Rect(0, 0, 120, 50)
    pn_text_oms = page.get_textbox(rect_pn_oms).rstrip().replace('\n','')
    if pn[:7] in pn_text: # file is drawing
    #         print('drw:',pn_text)
        if pn in pn_text:
            page.draw_rect(rect_pn, width=1.5, color=(0, 0, 1))
            pn_result = re.findall(pn[:11],pn_text)[-1]
        else:
            red_line = True
            pn_result = 'PN not match ' + '(' + pn_text + ')'
            page.draw_rect(rect_pn, width=1.5, color=(1, 0, 0))

        rect_rev = Rect(page.rect.width-60, page.rect.height-64,\
                                page.rect.width-31, page.rect.height-48)
        rev_text = page.get_textbox(rect_rev)
        if pn_rev in rev_text\
        and '+' not in rev_text:
            rev_result = re.findall(pn_rev,rev_text)[-1]
            page.draw_rect(rect_rev, width=1.5, color=(0, 0, 1))
        else:
            rev_result = '(' + rev_text + ')' + ' Not match Rev'
            page.draw_rect(rect_rev, width=1.5, color=(1, 0, 0))
            red_line = True

        rect_des = Rect(page.rect.width-295, page.rect.height-125,\
            page.rect.width-32, page.rect.height-75)
        des_text = page.get_textbox(rect_des).rstrip().replace('\n','')

    #         print(des_text,'_____\n',des_text.replace('\n',''))
        if des_input == '':
            des_result = 'Des. in Drawing ' + '(' + des_text + ')'
            page.draw_rect(rect_des, width=1.5, color=(0, 0, 1))
        elif des_input == des_text:
            des_result = des_text
            page.draw_rect(rect_des, width=1.5, color=(0, 0, 1))
        else:
            red_line = True
            des_result = 'Des not match ' + '(' + des_text + ')'
            page.draw_rect(rect_des, width=1.5, color=(1, 0, 0))
    elif pn[:7] in pn_text_oms: # file is OMS
    #         print('oms:',pn_text_oms)
    #                             rect_pn_oms_test = Rect(-50,-100,page.rect.width/2,page.rect.height/2)
        oms_pn = page.search_for(pn[:7],hit_max=16, quads=False, flags=1)
    #                             r = fitz.Rect(oms_pn[0])
        highlight = page.add_highlight_annot(oms_pn[0])
        highlight.set_colors({"stroke":(0, 0, 1)})
        highlight.update()
        if 'x' in pn_text_oms.lower():
            pn_result = pn_text_oms
        else:
            pn_result = 'Dash number need to change:' + '(' + pn_text_oms + ')'
    #                             page.draw_rect(oms_pn[0], width=1.5, color=(0, 0, 1))
        rec_rev_oms = Rect(page.rect.width-150, 50,\
                page.rect.width-20,100)
        rev_text_oms = page.get_textbox(rec_rev_oms)
        if pn_rev in rev_text_oms:
            rev_result = re.findall(pn_rev,rev_text_oms)[-1]
        else:
            red_line = True
            rev_result = '(' + rev_text_oms + ')' + ' Not match Rev'
    if key_check != '':
        key_check_result = []
        for i in range(len(doc)): #find key check in all page
    #         print('page', i)
            page: Page = doc[i]
            text = page.get_text()
            key_found = []
            for key in key_check:
                search_founds = page.search_for(key)
                if len(search_founds) > 0:
                    key_found.append(key)
                    red_line = True
                    for item in search_founds:
                        highlight = page.add_highlight_annot(item)
                        highlight.set_colors({"stroke":(1, 0, 0)})
                        highlight.update()
            if len(key_found) > 0:
    #             print(key_found)
                key_check_result.append(f'page {i+1} found key(s): {"/".join(key_found)}')
        if len(key_check_result) > 0:

            key_check_result = '\n'.join(key_check_result)
        else: key_check_result = ''
    if red_line:
        viz_name = os.path.join(file_path, file[:-4] + "________Redline.pdf" )
    else:
        viz_name = os.path.join(file_path, file[:-4] + "_SelfQC.pdf" )
    doc.save(viz_name)
    return file_name_result,pn_result,rev_result,des_result,key_check_result
```
